```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Calculate the total number of schools
total_schools = school_data['School ID'].count()
total_schools

```

```python
# Calculate the total number of students
total_students = student_data['Student ID'].count()
total_students
```

```python
#Calculate the total budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
#Calculate the average math score
average_math= student_data['math_score'].mean()
average_math
```

```python
#Calculate the average reading score
average_reading = student_data['reading_score'].mean()
average_reading
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
# 1 define the condition
# 2 use loc to extract the right data
#df[percent] = (df['column_name'] / df['column_name'].sum()) * 100
#math_passing['passing']=student_data['math_score']>=70
#math_passing['passing'].count

math_pass = student_data.loc[student_data['math_score'] >=70,:]
math_pass_total =math_pass.count()
#math_pass_total
total_math_scores = student_data['math_score'].count()
math_pass_rate = (math_pass_total/total_math_scores)*100
math_pass_rate['math_score']
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
reading_pass = student_data.loc[student_data['reading_score'] >=70,:]
reading_pass_total =reading_pass.count()
total_reading_scores = student_data['reading_score'].count()
reading_pass_rate = (reading_pass_total/total_reading_scores)*100
reading_pass_rate['reading_score']
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_pass = student_data.loc[(student_data['math_score'] >=70)&(student_data['reading_score'] >=70),:]
overall_pass_total = overall_pass.count()
overall_pass_rate=(overall_pass_total/total_students)*100
overall_pass_rate['Student ID']
```

```python
#Create a dataframe to hold the above results
summary_df=pd.DataFrame(
    {'Total Schools':[total_schools],
     'Total Students':[total_students],
     'Total Budget':[total_budget],
     'Average Math Score': [average_math],
     'Average Reading Score': [average_reading],
     '% Passing Math':[math_pass_rate['math_score']],
     '% Passing Reading': [reading_pass_rate['reading_score']],
     '% Overall Passing':[overall_pass_rate['Student ID']]
        
    }
)
summary_df
```

```python
overview_school_data = school_data_complete.groupby(['school_name', "type"])
overview_school_data
```

```python
#Total School Budget
total_budget_set = overview_school_data["budget"].mean()
total_budget_set
```

```python
#Per Student Budget
budget_per_student =  (total_budget_set/ total_students_set)
budget_per_student
```

```python
#Average Math Score
average_math_set = overview_school_data["math_score"].mean()
average_math_set
```

```python
#Average Reading Score
average_reading_set = overview_school_data["reading_score"].mean()
average_reading_set
```

```python
#% Passing Math
math_pass_percent = school_data_complete.query('math_score >70')["School ID"].count()/ total_students_set
math_pass_percent
```

```python
reading_pass_percent = school_data_complete.query('reading_score >70')["School ID"].count()/ total_students_set 
reading_pass_percent
```

```python
overall_pass_set = ((average_math_set + average_reading_set)/2)
overall_pass_set
```

```python

  
# Converting a GroupBy object into a DataFrame
overview_school_data_df = pd.DataFrame({"Total Students":total_students_set,
                                       "Total School Budget": total_budget_set, 
                                       "Per Student Budget":budget_per_student,
                                       "Average Math Score": average_math_set,
                                       "Average Reading Score": average_reading_set,   
                                       "% Passing Math":math_pass_percent,
                                       "% Passing Reading":reading_pass_percent,
                                        "%Overall Passing Rate":overall_pass_set         
                                    
                                      
})

overview_school_data_df = overview_school_data_df[[  "Total Students",
    "Total School Budget",
    "Per Student Budget",
    "Average Math Score",
    "Average Reading Score",
     "% Passing Math",   
     "% Passing Reading",                                             
    "%Overall Passing Rate"]]



overview_school_data_df["Total School Budget"] = overview_school_data_df["Total Students"].map("${:,.2f}".format)
overview_school_data_df["Per Student Budget"] = overview_school_data_df["Per Student Budget"].map("${:,.2f}".format)

overview_school_data_df.head(10)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

```python
grade_nine = school_data_complete[(school_data_complete["grade"] == "9th")]
grade_nine_scores = grade_nine.groupby(["school_name"]).mean()["math_score"]

grade_ten = school_data_complete[(school_data_complete["grade"] == "10th")]
grade_ten_scores = grade_ten.groupby(["school_name"]).mean()["math_score"]

grade_eleven = school_data_complete[(school_data_complete["grade"] == "11th")]
grade_eleven_scores = grade_eleven.groupby(["school_name"]).mean()["math_score"]

grade_twelve = school_data_complete[(school_data_complete["grade"] == "12th")]
grade_twelve_scores = grade_twelve.groupby(["school_name"]).mean()["math_score"]


math_by_grade_df = pd.DataFrame({"9th": grade_nine_scores,
                                "10th":grade_ten_scores,
                                "11th":grade_eleven_scores,
                                "12th":grade_twelve_scores
                                    }
                                 )


math_by_grade_df.index.name = None

math_by_grade_df =math_by_grade_df [["9th", "10th", "11th", "12th"]]


math_by_grade_df
```
